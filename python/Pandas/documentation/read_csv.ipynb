```python
# importando a biblioteca pandas
import pandas as pd
```

> Observação: Quando for efetuar a leitura de um csv no padas observar o separador do arquivo se é vírgula ou ponto e virgua  e assim por diante. Outro ponto interessante é sempre definir o encoding com UTF-8. Outro argumento é definir em que linha inicial seu cabeçalho para isso utilizamos o método header e especificamos a linha onde está o cabeçalho.


```python
# Abrindo um arquivo csv 
df=pd.read_csv('banco_1.csv', encoding='UTF-8', sep=';', header=0)
df
```

> Mostrar o shape dos dados, ou seja, Numero de Linhas e coluna do Data Frame

```python
df.shape
```

> Selecionando colunas dentro do meu Data Frame para isso utilizo o argumento usecols passando uma lista com todas  as colunas que quero como modelo do meu Data Frame. Posso ainda determinar a quantidade de linhas a serem retornadas na leitura do meu arquivo para isso vou usar o argumento rows 

```python
df=pd.read_csv('banco_1.csv', encoding='UTF-8', sep=';', header=0, usecols=['AddressID','City','ModifiedDate'], nrows=100)
df
```
